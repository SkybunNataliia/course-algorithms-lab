```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import math
```

```python
fig, ax = plt.subplots()
x = np.linspace (1, 20, num=200)
functions = [float, math.log, lambda n: n*math.log(n), lambda n: n*n, lambda n: n*n*n, math.gamma, lambda n: math.pow(n,n)]
labels = ["n", "log_e(n)", "nlog(n)", "n^2", "n^3", "gamma(n)", "n^n"]

for f, l in zip(functions, labels):
    y = list(map(f, x))
    ax.plot(x, y, label=l)
    ax.text(x[-1], y[-1], l)

ax.set_xscale('log')
ax.set_ylim(0.1, 80)

ax.set_xlabel("n")
ax.set_ylabel("f(n)")

ax.legend(loc="upper left")

plt.show()
```
